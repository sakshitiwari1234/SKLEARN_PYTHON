## What is column transformer ?

Column transformer is a class which allows to put class in all the columns at once in a dataset to convert the columns into disearable data encoding at once.

```python
from sklearn.compose import ColumnTransformer as tf
```

```python
transformer = ColumnTransformer(
    transformers=[
        ("tf1",SimpleImputer(), [['fever']]),
        ("tf2", OrdinalEncoder(categories = [["mild", 'strong']]),['cough'] ),
        ("tf3", OneHotEncoder(drop = "first"),['gender', 'city']),




    ],
    remainder = "passthrough"
)


# here remiander keeping passthorugh means that if a table has 10 colmns and wwe are defining only 5 transformers then, it won't disturb the rest of them at all and will be remain same. thankyou so much i like studying this concept vert much and hope to you too.
```
