### Train Test Split

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

### what is standard scalar?

it involves the features in a dataset so that they have a mean of zero and a standard deviation of one.

new mean  = object - mean / deviation.
