What is One-Hot Encoding?
Definition:
One-hot encoding is a technique used to convert categorical data into a numerical format that machine learning models can work with. It represents each unique category in the data as a binary vector.

How Does It Work?
Each category gets its own column.
The value is marked as 1 in the column corresponding to the category and 0 elsewhere.
Example:
Imagine we have a categorical feature Color with values:
["Red", "Green", "Blue"]

After one-hot encoding, the categories will be represented like this:

Color	Red	Green	Blue
Red	1	0	0
Green	0	1	0
Blue	0	0	1
Each row is a one-hot vector, where only one position is "hot" (has 1), and the rest are "cold" (have 0s).

Why is One-Hot Encoding Used?
For Numerical Representation:

Machine learning models work better with numbers. Since categories like "Red" or "Blue" are not inherently numerical, one-hot encoding turns them into numbers in a way that doesn't imply any order or relationship.
Avoiding Misinterpretation:

Simply assigning numbers to categories (e.g., Red = 1, Green = 2, Blue = 3) might confuse the model into thinking theres an ordinal relationship (e.g., Green > Red).
Preserving Categorical Information:

It ensures that the model understands the categories are distinct and unrelated.
Why Can't We Use Labels Instead?
If we assign labels, like:
Red = 1, Green = 2, Blue = 3
The model might think Blue is greater than Green and Red, which is incorrect.
One-hot encoding eliminates this by treating each category equally.
