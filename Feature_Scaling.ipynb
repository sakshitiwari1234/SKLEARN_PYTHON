### What is feature scaling

feature scaling is a data preprocessing technique used to transform the value of features or variables in a dataset to a similar scale.

The purpose is to ensure that all features constribute equally to the model and to avoid the domination of features with larger values.

####  there are two types of scaling measurement used in scikit-learn.

1. Normalization.
2. Standardization.
